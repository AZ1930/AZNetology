Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из sРоссии - ставим other
все остальные варианты берем из traffic_source без изменений

```python
# Задание 1.
import pandas as pd
import numpy as np
```

```python
log = pd.read_csv('visit_log.csv', sep = ';')
log.head()
log.loc[(log.traffic_source == 'yandex')|(log.traffic_source == 'google'), 'source_type'] = 'organic'

log.loc[((log.traffic_source == 'paid')|(log.traffic_source == 'mail')) & (log.region != 'Russia'), 'source_type'] = 'others'
log.loc[((log.traffic_source == 'paid')|(log.traffic_source == 'mail')) & (log.region == 'Russia'), 'source_type'] = 'ad'
log.head(20)
```

```python
import pandas as pd
import numpy as np
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd = rzd[['client_id', 'rzd_revenue']]
rzd_grouped = rzd.groupby('client_id')#.count()
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto = auto[['client_id', 'auto_revenue']]
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air = air[['client_id', 'air_revenue']]
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base = client_base[['client_id', 'address']]
client_base
```

```python
mergedata1 = client_base.merge(rzd, how='left', on = 'client_id')
mergedata2 = mergedata1.merge(air, how='left', on = 'client_id' )
mergedata3 = mergedata2.merge(auto, how='left', on = 'client_id')
mergedata4 = mergedata3.drop(columns='address')
mergedata4 # без адресов
```
