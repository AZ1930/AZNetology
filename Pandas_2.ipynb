Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd 
```

```python
# Задание 1.
stats = pd.read_csv('ratings_2.csv',sep = ',', header=0)
stats.head()
```

```python
def class_column(row):
    if row.rating>=4.5:
        return 'высокий рейтинг'
    elif 2<row.rating<4.5:
        return 'средний рейтинг'
    elif row.rating<=2:
        return 'низкий рейтинг'
        
stats['class']=stats.apply(class_column, axis=1)
stats.head()
```

```python
stats['class'].value_counts().head()
```

```python
# Задание 2.
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
stats = pd.read_csv('keywords.csv')
stats.loc[125:130]
```

```python
def region (row):
    for key, value in geo_data.items():
        for i in value:
            if i in row['keyword']:
                return key
    return 'undefined'
print('geo -', geo_data)
stats['region'] = stats.apply(region, axis = 1) 
stats[(stats['region'] == 'Дальний Восток')].head(10)
#stats['region'].value_counts().head() # для проверки всех регионов
```
