Задача 1.
Дана переменная, в которой хранится словарь, 
содержащий гео-метки для каждого пользователя 
(пример структуры данных приведен ниже). 
Вам необходимо написать программу, 
которая выведет на экран множество уникальных гео-меток всех пользователей.

Пример работы программы:

ids = {‘user1’: [213, 213, 213, 15, 213],
‘user2’: [54, 54, 119, 119, 119],
‘user3’: [213, 98, 98, 35]}
Результат: {98, 35, 15, 213, 54, 119}

***    Использовать множества.
Тут очень пригодится использование множеств. 
Вы можете просто перебирать циклом значения структуры данных и
сразу формировать результат через set, все обойдётся одним циклом;


```python
# Задача 1.
ids = {'user1': [213, 213, 213, 15, 213],
'user2': [54, 54, 119, 119, 119],
'user3': [213, 98, 98, 35]}
ids_res = {}
for element in ids:
    ids_res = set(ids[element]).union(ids_res)
print('Результат:',(ids_res))
    
```
