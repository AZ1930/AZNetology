```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
# Функция 1
def name_by_document_number(doc_number):
    """ 1.Пользователь по команде "p" может узнать владельца документа по его номеру"""
    for i in range(len(documents)):
        k = documents[i]['number']
        if doc_number == k:
            res = documents[i]['name']
            break
        else:
            res = 'Документ не найден в базе'
    return res
# Функция 2
def shelf_by_document_number(doc_number_2):
    """2.Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится"""
    for shelf in directories:
        if doc_number_2 in directories.get(shelf):
            res_2 = shelf
            break
        else:
            res_2 = 'Документ не найден в базе'
    return res_2
# Функция 3
def full_information():
    """3.Пользователь по команде "l" может увидеть полную информацию по всем документам"""
    for p in range(len(documents)):
        k = documents[p]
        for key, v in directories.items():
            for i in range(len(v)):
                if k['number'] == str(v[i]):
                    print('№: ',k['number'],', ', 'тип: ', k['type'],', владелец: ',k['name'],', полка хранения: ',key)
                
    return k['number']
# Функция 4
def add_shelf(new_shelf):
    """4.Пользователь по команде "ads" может добавить новую полку"""
    list_1 = list(directories)
    print(list_1)
    if new_shelf in directories.keys():
        print('Такая полка уже существует.Текущий перечень полок:',list(directories.keys()))
    else:
        directories[new_shelf] = []
        print('Полка добавлена. Текущий перечень полок:',list(directories.keys()))
    return new_shelf
# Функция 5
def delete_shelf():
    """Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)"""
    if del_shelf in directories.keys() and directories[del_shelf] != []:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:',list(directories.keys()))    
    
    elif del_shelf in directories.keys() and directories[del_shelf] == []:
        del(directories[del_shelf])
        print('Полка удалена. Текущий перечень полок:',list(directories.keys()))
        
    elif del_shelf not in directories.keys():
        print('Такой полки не существует. Текущий перечень полок: ',list(directories.keys()))
    return del_shelf
# Функция ввода данных пользователем
def user_input():
    """Ввод данных пользователем"""
    if command == 'p':
        doc_number = (input('Введите номер документа (определим владельца): '))
        name_by_document_number(doc_number)
        print('Владелец документа:', name_by_document_number(doc_number))  
    elif command == 's':
        doc_number_2 = (input('Введите номер документа (определим полку): '))
        shelf_by_document_number(doc_number_2)
        print('Документ лежит на полке:', shelf_by_document_number(doc_number_2))
    elif command == 'l':
        print('Ваше инфо:')
        full_information()
    elif command == 'ads':
        new_shelf = input('Введите номер полки для добавления: ')
        add_shelf(new_shelf)
    elif command == 'ds':
        del_shelf = str(input('Введите номер полки для удаления: '))
        delete_shelf()
        
    elif command == 'q':
        print('Выход из программы')
    else:
        print('Не та команда ')
command = ''    
command = input('Введите команду: ')
user_input()

```
