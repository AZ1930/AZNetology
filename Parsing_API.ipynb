```python
# Задание 1.
def get_all_links(url, query, pages):
    all_refs = []
    params = {
        'q': query
    }
    for i in range(1, pages + 1):
        params['page'] = i
        print('params=',params['page'])
        res = requests.get(URL, params)
        print('res=',res)
        time.sleep(0.3)
        soup = BeautifulSoup(res.text)
        # Дата
        date_all = soup.find_all('span', class_='tm-article-snippet__datetime-published')
        date_habr = list(map(lambda z: z.find('time').get('datetime'), date_all))
        print('date_habr=', date_habr )
        # Заголовок
        new_blocks = soup.find_all('h2', class_='tm-article-snippet__title')
        headlines = list(map(lambda p: p.find('span'), new_blocks))
        title = headlines
        print('zagolovok=',title)
        # Ссылка
        articles_intro = list(map(lambda x:x.find('a',class_='tm-article-snippet__title-link'or'data-article-link=""'), new_blocks))
        a_list = list(map(lambda x: x.get('href'), articles_intro))
        all_refs = list(map(lambda x: 'https://www.habr.com' + x, a_list))
        print('ssilka=', all_refs)
    return all_refs


def get_kom_news(links):
    kom_news = pd.DataFrame()
    for link in all_links:
        soup = BeautifulSoup(requests.get(link).text)
        date = pd.to_datetime(soup.find('span', class_='tm-article-snippet__datetime-published').find('time').get('datetime'), dayfirst=True).date()
        if soup.find('h2', class_='tm-article-snippet__title').text:
            print('0000')
        else:
            title = soup.find('h2', class_='tm-article-snippet__title').text
            
        print('title =',title)
        text = soup.find('a', class_='tm-article-snippet__title-link').text
        row = {'date': date, 'title': title, 'text': text}
        kom_news = pd.concat([kom_news, pd.DataFrame([row])])  
    return kom_news
```

```python
URL = 'https://habr.com/ru/search/'
all_links = get_all_links(URL, 'python', 3)
all_links
kom_news = get_kom_news(all_links)
kom_news
```
